```python
import pandas as pd
# Carga el dataset de trabajo
netflix = pd.read_csv("../Datos/netflix_merged.csv",index_col=0)
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
year_filter = netflix["release_year"]==2021
type_filter = netflix["type"]=="Movie"
netflix[year_filter & type_filter]
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
duration_filter = netflix["duration"] == "90 min"
netflix.loc[duration_filter,["title","duration"]]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
year_filter = netflix["release_year"] > 2018
country_filter = netflix["country"] == "India"
netflix.loc[duration_filter,["title","duration"]]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
clasificaciones = ["TV-14","R"]
rating_filter = netflix["rating"].isin(clasificaciones)
netflix.loc[rating_filter,["title"]]
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
directores = ['Christopher Nolan','Martin Scorsese']
director_filter = netflix["director"].isin(directores)
netflix.loc[director_filter,["title"]]
```

```python
netflix.loc[director_filter,["title","director"]].groupby("director").count()
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
countries = ['United States','United Kingdom','Canada']
country_filter = netflix["country"].isin(countries)
netflix.loc[country_filter,["title"]]
```

```python
language_filter = netflix["Language"].str.contains("/")
netflix.loc[country_filter & language_filter,["Language"]].count()
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
years_filter = netflix["release_year"].between(2015,2020,inclusive="both")
netflix.loc[years_filter,["title"]]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
score_filter = netflix["IMDB Score"].between(6.5,8,inclusive="both")
duration_filter = netflix["duration"].isna()
netflix.loc[score_filter & duration_filter,["title"]]
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
duration_filter = netflix.duration.isin(["90 min","150 min"])
netflix.loc[duration_filter,["title"]]
```

```python
netflix_movies = netflix[netflix["type"] == "Movie"]
netflix_movies["duration"] = netflix_movies["duration"].str.replace(" min","").astype(float)
duration_filter = netflix_movies["duration"].between(90,150,inclusive="both")
netflix_movies.loc[duration_filter,["title"]]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
contains_love = netflix["title"].str.contains("love",case=False)
type_filter_show = netflix["type"] == "TV Show"
netflix.loc[contains_love & type_filter_show,["title"]]
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
surname_johnson = netflix["director"].str.contains("Johnson",case=False).fillna(False)
netflix.loc[surname_johnson,["title"]]
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
documentary_filter = netflix["listed_in"].str.contains("docu",case=False)
tyson_documentary = netflix["title"].str.contains("tyson",case=False)
netflix.loc[documentary_filter & tyson_documentary,["title"]]
```

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').

```python
netflix.loc[netflix["director"].isna(),["title"]]
```

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
title_isnotnan = ~netflix["title"].isna()
country_isnotnan = ~netflix["country"].isna()
IMDB_isnotnan = ~netflix["IMDB Score"].isna()
netflix.loc[title_isnotnan & country_isnotnan & IMDB_isnotnan,["title"]]
```
